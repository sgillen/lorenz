```python
from seagul.rl.ars.ars_pool import ars
import torch
torch.set_default_dtype(torch.double)
import seagul.envs
import matplotlib.pyplot as plt
from seagul.nn import MLP
import gym
import dill as pickle

torch.set_default_dtype(torch.float64)
env_name = "Humanoid-v2"
env = gym.make(env_name)
in_size = env.observation_space.shape[0]
out_size = env.action_space.shape[0]

policy = MLP(in_size, out_size,0,0,bias=False)
```

```python
import time
start = time.time()
policy, r_hist = ars(env_name, policy, 250, n_workers=16, n_delta=64, n_top=32)
print(time.time() - start)

plt.plot(r_hist)
plt.show()

#env = gym.make(env_name)
#state_hist, act_hist, returns = do_rollout(env_name, policy)
```
